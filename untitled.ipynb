```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 1000)

# Compute the corresponding real and imaginary parts for e^(ix) and e^(-ix)
real_part = np.cos(x)
imaginary_part = np.sin(x)

# Create the first plot for e^(ix)
fig1 = plt.figure()
ax = fig1.add_subplot(111, projection='3d')
ax.plot(x, real_part, imaginary_part, label='$e^{ix}$')
ax.set_xlabel('x')
ax.set_ylabel('Real Part')
ax.set_zlabel('Imaginary Part')
ax.set_title('3D Visualization of $e^{ix}$')
ax.legend()
ax.view_init(elev=20, azim=60)

# Show the first plot
plt.show()

# Create the second plot for e^(-ix)
fig2 = plt.figure()
bx = fig2.add_subplot(111, projection='3d')
bx.plot(x, real_part, imaginary_part, label='$e^{-ix}$')
bx.set_xlabel('x')
bx.set_ylabel('Real Part')
bx.set_zlabel('Imaginary Part')
bx.set_title('3D Visualization of $e^{-ix}$')
bx.legend()
bx.view_init(elev=0, azim=0)

# Show the second plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter sinyal analog
A = 3  # Amplitudo (volt)
f = 2  # Frekuensi (Hz)
t = np.linspace(0, 1, 500)  # Waktu dari 0 hingga 1 detik (500 titik)

# Sinyal analog: V(t) = A * sin(2*pi*f*t)
V_analog = A * np.sin(2 * np.pi * f * t)

# Parameter sampling dan kuantisasi
sampling_rate = 8  # Sampling rate (Hz)
bit_depth = 4  # Bit depth
num_levels = 2**bit_depth  # Jumlah level kuantisasi
adc_range = (0, 3)  # Rentang ADC (volt)
adc_resolution = (adc_range[1] - adc_range[0]) / num_levels  # Resolusi ADC

# Waktu sampling
t_sampled = np.arange(0, 1 + 1/sampling_rate, 1/sampling_rate)

# Sinyal analog pada waktu sampling
V_sampled = A * np.sin(2 * np.pi * f * t_sampled)

# Kuantisasi sinyal sampled
V_quantized = np.round((V_sampled - adc_range[0]) / adc_resolution) * adc_resolution + adc_range[0]

# Plot sinyal analog
plt.figure(figsize=(10, 6))
plt.plot(t, V_analog, label="Sinyal Analog", color="blue")

# Plot sinyal sampled dan quantized
plt.stem(t_sampled, V_quantized, linefmt="r-", markerfmt="ro", basefmt=" ", label="Sinyal Digital (Quantized)")

# Format plot
plt.title("Visualisasi Gelombang Analog dan Digital")
plt.xlabel("Waktu (detik)")
plt.ylabel("Tegangan (Volt)")
plt.grid(True)
plt.legend()
plt.ylim(-3.5, 3.5)  # Batas sumbu y
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter sinyal analog
amplitude = 2  # Amplitudo sinyal (volt)
frequency = 1  # Frekuensi sinyal (Hz)

# Parameter ADC
sampling_rate = 4  # Sampling rate (Hz)
bit_depth = 3  # Bit depth
adc_range = (0, 2)  # Rentang tegangan ADC (volt)

# Rentang sinyal analog
analog_range = (-amplitude, amplitude)

# Waktu sampling (0.00, 0.25, 0.50, 0.75, 1.00)
t = np.array([0.00, 0.25, 0.50, 0.75, 1.00])

# 1. Generate sinyal analog (gelombang sinus)
def analog_signal(t):
    return amplitude * np.sin(2 * np.pi * frequency * t)

# 2. Normalisasi sinyal ke rentang ADC
def normalize_signal(signal, analog_range, adc_range):
    analog_min, analog_max = analog_range
    adc_min, adc_max = adc_range
    # Normalisasi ke rentang 0 hingga (adc_max - adc_min)
    normalized = ((signal - analog_min) / (analog_max - analog_min)) * (adc_max - adc_min)
    return normalized

# 3. Kuantisasi sinyal
def quantize_signal(signal, bit_depth, adc_range):
    adc_min, adc_max = adc_range
    num_levels = 2 ** bit_depth
    resolution = (adc_max - adc_min) / num_levels
    # Hitung level kuantisasi
    quantized_levels = np.floor(signal / resolution).astype(int)
    # Pastikan level tidak melebihi jumlah level yang tersedia
    quantized_levels = np.clip(quantized_levels, 0, num_levels - 1)
    return quantized_levels, resolution

# 4. Encoding ke bentuk biner
def encode_signal(quantized_levels, bit_depth):
    binary_format = f'0{bit_depth}b'  # Format biner dengan leading zero
    binary_signal = [format(level, binary_format) for level in quantized_levels]
    return binary_signal

# Proses konversi
analog_signal_values = analog_signal(t)  # 1. Ambil sampel sinyal analog
normalized_signal_values = normalize_signal(analog_signal_values, analog_range, adc_range)  # 2. Normalisasi
quantized_levels, resolution = quantize_signal(normalized_signal_values, bit_depth, adc_range)  # 3. Kuantisasi
binary_signal = encode_signal(quantized_levels, bit_depth)  # 4. Encoding ke biner

# Output hasil
print("Waktu (t):", t)
print("Sinyal Analog (V(t)):", analog_signal_values)
print("Sinyal Normalisasi (V_norm(t)):", normalized_signal_values)
print("Level Kuantisasi:", quantized_levels)
print("Sinyal Digital (Biner):", binary_signal)

# Visualisasi
plt.figure(figsize=(10, 6))

# Plot sinyal analog
plt.subplot(2, 1, 1)
plt.plot(t, analog_signal_values, 'b-', label='Sinyal Analog', marker='o')
plt.title('Sinyal Analog vs Digital')
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo (V)')
plt.grid(True)
plt.legend()

# Plot sinyal digital
plt.subplot(2, 1, 2)
plt.step(t, quantized_levels, 'r-', label='Sinyal Digital', where='post', marker='o')
plt.title('Sinyal Digital (Kuantisasi)')
plt.xlabel('Waktu (t)')
plt.ylabel('Level Kuantisasi')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
