Buat sinyal baru y5 yang merupakan kombinasi dari y1 dan y2 yaitu, y5 = y1 + y2
Plot y5 dan diskusikan bagaimana kombinasi dua gelombang sinus/cosinus mempengaruhi bentuk gelombang yang dihasilkan.

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat array t dari 0 hingga 2 dengan langkah 0.0001
t = np.arange(0, 2, 0.0001)

# Mendefinisikan sinyal
y1 = 2 * np.sin(2 * np.pi * 3 * t)
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi / 4)

# Menentukan sinyal baru hasil penjumlahan y1 + y2
y5 = y1 + y2

# Menentukan rentang y yang sesuai
y_min, y_max = y5.min(), y5.max()

# Plot sinyal y5
plt.figure(figsize=(10, 4))
plt.plot(t, y5, 'c', label='y5 = y1 + y2')
plt.title("y5 = y1 + y2")
plt.ylim(y_min, y_max)
plt.grid()
plt.legend()
plt.show()
```

### Bagaimana Kombinasi Dua Gelombang Sinus/Cosinus Mempengaruhi Bentuk Gelombang?

### Jawaban
y1 adalah gelombang sinus dan y2 adalah gelombang cosinus dengan pergeseran fase /4, hasilnya adalah pola gelombang baru yang tidak murni sinusoidal.
#### Pengaruh Amplitudo
y1 memiliki amplitudo 2, sedangkan y2 memiliki amplitudo 1.Karena amplitudo y1 lebih besar, kontribusinya terhadap bentuk gelombang y5 lebih dominan.
#### Pengaruh Frekuensi
y1 memiliki frekuensi 3 Hz , sementara y2 memiliki frekuensi 4 Hz. Perbedaan frekuensi ini menyebabkan pola gelombang yang lebih kompleks, di mana puncak dan lembah dari y1 dan y2 tidak selalu sejajar.

Berdasarkan penjumlahan dua gelombang tersebut hasilnya bukan sinus atau cosinus murni, tetapi kombinasi dari keduanya dengan variasi amplitudo dan bentuk yang lebih kompleks
